```python
import json
import requests
import pandas as pd
from config import api_key_weather
from citipy import citipy
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# api.openweathermap.org/data/2.5/find?q=London&units=metric
# http://api.openweathermap.org/data/2.5/weather?"
url_weather = "http://api.openweathermap.org/data/2.5/find?"
city = 'Mexico'

query_url = url_weather + 'appid=' + api_key_weather + '&q=' + city + '&units=imperial'
```

```python
 # This list is for the info obtained from the City import library
cities = {'City name':[]}
count = 0
# I created a new dictionary to save the data from the search request
weatherdf={'City id':[], 'City name':[], 
                       'Latitude':[],'Temperature in F':[], 'Humidity':[], 'Cloudiness':[], 
                       'Wind Speed (mph)':[]}


### There will need to be 500 elements in the dictionary, I used a while to check up 
while count < 501:
    # Generated a random number for Latitud and Longitud
    city = citipy.nearest_city(random.uniform(-90,90), random.uniform(-180,180)) 
    # Then checked if it was repeated
    if city.city_name not in cities['City name']:
        # Search for the city in the library 
        cities['City name'].append(city.city_name)
        query_url = url_weather + 'appid=' + api_key_weather + '&q=' + city.city_name + '&units=imperial'
        weather_response = requests.get(query_url)
        weather_json = weather_response.json()
        if (weather_json['list'] != []) & (weather_json['cod'] != ['400']) :
            weatherdf['City id'].append(weather_json['list'][0]['id'])
            weatherdf['City name'].append(weather_json['list'][0]['name'] )
            weatherdf['Latitude'].append(weather_json['list'][0]['coord']['lat'] )
            weatherdf['Temperature in F'].append(weather_json['list'][0]['main']['temp'])
            weatherdf['Humidity'].append(weather_json['list'][0]['main']['humidity'])
            weatherdf['Cloudiness'].append(weather_json['list'][0]['clouds']["all"] )
            weatherdf['Wind Speed (mph)'].append(weather_json['list'][0]['wind']["speed"])
            print(f"########{count}. City:{weatherdf['City name'][count]}  Country: {city.country_code} /Ok")
            count += 1
        else:
              print('#### Without info -/### Searching for another city.')
    else: print('#### Repeated -/### Searching for another city.')

```

Crear las clases, además de definir el código para imprimir la gráfica!

```python
weathercity = pd.DataFrame(weatherdf)
```

```python
weathercity.to_csv('Weather.csv',index=False)
```

```python
title1 = 'Temperature (F) vs. Latitude'
title2 = 'Humidity (%) vs. Latitude'
title3 = 'Cloudiness (%) vs. Latitude'
title4 = 'Wind Speed (mph) vs. Latitude'

fig, axes = plt.subplots(1, 4, figsize=(15,3))

axes[0].scatter(weathercity['Latitude'],weathercity['Temperature in F'])
axes[0].set_title(title1)


axes[1].scatter(weathercity['Latitude'],weathercity['Humidity'])
axes[1].set_title(title2)

axes[2].scatter(weathercity['Latitude'],weathercity['Cloudiness'])
axes[2].set_title(title3)

axes[3].scatter(weathercity['Latitude'],weathercity['Wind Speed (mph)'])
axes[3].set_title(title4)
# axes[1].step(n, n**2, lw=2)
# axes[1].set_title("step")

# axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
# axes[2].set_title("bar")

# axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
# axes[3].set_title("fill_between");
```

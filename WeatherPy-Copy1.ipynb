```python
class WeatherPy():
    
    
    
    def __init__ (self) :
        pass
    
    def weatherinfo(self):
        cities = {'City name':[]}
        count = 0
        weatherdf={'City id':[], 'City name':[], 
                  'Latitude':[],'Temperature in F':[], 'Humidity':[], 'Cloudiness':[], 
                  'Wind Speed (mph)':[]}

        while count < 501:
            city = citipy.nearest_city(random.uniform(-90,90), random.uniform(-180,180)) 
            if city.city_name not in cities['City name']:
                cities['City name'].append(city.city_name)
                query_url = url_weather + 'appid=' + api_key_weather + '&q=' + city.city_name + '&units=imperial'
                weather_response = requests.get(query_url)
                weather_json = weather_response.json()
                if (weather_json['list'] != []) | (weather_json['cod'] != ['400']) :
                    weatherdf['City id'].append(weather_json['list'][0]['id'])
                    weatherdf['City name'].append(weather_json['list'][0]['name'] )
                    weatherdf['Latitude'].append(weather_json['list'][0]['coord']['lat'] )
                    weatherdf['Temperature in F'].append(weather_json['list'][0]['main']['temp'])
                    weatherdf['Humidity'].append(weather_json['list'][0]['main']['humidity'])
                    weatherdf['Cloudiness'].append(weather_json['list'][0]['clouds']["all"] )
                    weatherdf['Wind Speed (mph)'].append(weather_json['list'][0]['wind']["speed"])
                    print(f"########{count}. City:  Country:{weatherdf['City name'][count]} {city.country_code} /Ok")
                    count += 1
                else:
                      print('#### Without info -/### Searching for another city.')
            else: print('#### Repeated -/### Searching for another city.')
        return pd.DataFrame(weatherdf)

    def tempvslat(self, weathercity):
        return sns.jointplot(x=weathercity['Latitude'],y=weathercity['Temperature in F'], 
                             height = 10,data=weathercity,kind='kde', color='crimson')
        
```
